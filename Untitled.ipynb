## CDF AND PDF

```python
iris_setosa=df.loc[df["Species"]=="Iris-setosa"];
counts, bin_edges=np.histogram(iris_setosa['PetalLengthCm'],bins=10,
                             density=True)
pdf=counts/sum(counts)
print(pdf)
print(bin_edges)
```

```python
iris_setosa
```

```python
cdf=np.cumsum(pdf);
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show();
```

### Mean, Median,Variance and STD-dev

```python
iris_virginica=df.loc[df["Species"]=="Iris-virginica"];
iris_versicolor=df.loc[df["Species"]=="Iris-versicolor"];
```

```python
print("Mean:")
print("iris_setosa",np.mean(iris_setosa["PetalLengthCm"]))
print("iris_versicolor",np.mean(iris_versicolor["PetalLengthCm"]))
print("iris_virginica",np.mean(iris_virginica["PetalLengthCm"]))
```

- petal length of versicolor and virginica are close 

- if there are outlier mean can change extremly 

- Mean tells central tendency 

- Varience - averge square distnce of each value 
- std = square root of variance -numerical representation of spread ,

```python
print("STD :")
print("iris_setosa",np.std(iris_setosa["PetalLengthCm"]))
print("iris_versicolor",np.std(iris_versicolor["PetalLengthCm"]))
print("iris_virginica",np.std(iris_virginica["PetalLengthCm"]))
```

```python
print("Median :")
print("iris_setosa",np.median(iris_setosa["PetalLengthCm"]))
print("iris_versicolor",np.median(iris_versicolor["PetalLengthCm"]))
print("iris_virginica",np.median(iris_virginica["PetalLengthCm"]))

# very less or no change with the outlier in the datset
# if more than 50% of the data is crupted then only the value of median get incorrect
```

### Percentile and Quantile 

```python
print("Quantile :")
print("iris_setosa",np.percentile(iris_setosa["PetalLengthCm"], np.arange(0,100,25)))
print("iris_versicolor",np.percentile(iris_versicolor["PetalLengthCm"], np.arange(0,100,25)))
print("iris_virginica",np.percentile(iris_virginica["PetalLengthCm"], np.arange(0,100,25)))
```

```python
print("Percentile :")
print("iris_setosa",np.percentile(iris_setosa["PetalLengthCm"],90))
print("iris_versicolor",np.percentile(iris_versicolor["PetalLengthCm"],90))
print("iris_virginica",np.percentile(iris_virginica["PetalLengthCm"],90))
```

```python
from statsmodels import robust
```

```python
#Median Absolute deviation
print("MAD :")
print("iris_setosa",robust.mad(iris_setosa["PetalLengthCm"]))
print("iris_versicolor",robust.mad(iris_versicolor["PetalLengthCm"]))
print("iris_virginica",robust.mad(iris_virginica["PetalLengthCm"]))
```

```python
# IQR - inter quatile range 
iqr1= np.percentile(iris_setosa["PetalLengthCm"],75)
iqr2= np.percentile(iris_setosa["PetalLengthCm"],25)
print(iqr1-iqr2)
```
